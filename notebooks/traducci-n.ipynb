```python
import pandas as pd
from googletrans import Translator

translator = Translator()


# Función para traducir texto
def traducir_texto(texto):
    try:
        print(f"Procesando reseña de {len(df)}")
        traduccion = translator.translate(texto, src='auto', dest='es').text
        return traduccion
    except Exception as e:
        return None

# Cargar el dataset original
archivo_entrada = "/kaggle/input/resenas-corregidas/reseas_corregidas.csv"  # Reemplázalo por la ruta de tu archivo
df = pd.read_csv(archivo_entrada)

df = df.dropna()
df.isnull().sum()
df['Rating'] = np.floor(df['Rating']).astype(int)

# Crear una nueva columna con las traducciones
df['Review_Traducida'] = df['Review'].apply(lambda texto: traducir_texto(texto) if pd.notnull(texto) else None)

# Filtrar las filas donde la traducción fue exitosa
df_traducido = df[df['Review_Traducida'].notnull()]
print('terminado')
```

```python
output = pd.DataFrame({'Review': df['Review_Traducida'], 'Rating': df['Rating']})
output = output.dropna()
output.to_csv('submission1.csv', index=False, encoding='utf-8')
print("Your submission was successfully saved!")
print(f"Traducción completada. Total reseñas traducidas: {len(df_traducido)}")
```
